```python
%matplotlib inline
from datascience import *
import numpy as np
import matplotlib.pyplot as plots
plots.style.use("fivethirtyeight")
import warnings
```

```python
table = Table()
```

```python
table = table.with_columns({"Name":["Alice", "Bob", " Charlie"], "Age":[25,30,20],"Salary":[50000,60000,45000]})
table
```

```python
def increase(m):
    return m*2
```

```python
families = Table().read_table(".csv/family_heights.csv")
```

```python
families.show(10)
```

```python
parent_averages = (families.column("father") + families.column("mother"))/2
```

```python
families = families.with_column("parent_average", parent_averages)
```

```python
families.show(10)
```

```python
heights = Table().with_columns("parent_average",parent_averages,"child", families.column("child"), "Sex", families.column("sex"))
```

```python
heights.scatter("parent_average", "child")
```

```python
heights.scatter("parent_average", "child")
plots.plot([67.7,67.5],[50,85],color="red",lw=2)
```

```python
nearby = heights.where("parent_average", are.between(67.5,68.5))
```

```python
nearby_mean = np.average(nearby.column('child'))
nearby_mean
```

```python
heights.scatter("parent_average", "child")
plots.plot([67.5,67.5],[50,85],color='red',lw=2)
plots.plot([68.5,68.5],[50,85],color='red',lw=2)
plots.scatter(68,nearby_mean,color='blue',s=50)
```

```python
def perdict(h):
    nearby = heights.where("parent_average", are.between(h-0.5,h+0.5))
    return np.average(nearby.column("child"))
perdict(68)
```

```python
heights.where("parent_average",68)
heights.show(5)
avg1 = heights.where("parent_average",68)
```

```python
avg2 = heights.where("parent_average",70)
```

```python
perdict(70)
```

```python
avgChild70 = np.average(avg2['child'])
```

```python
avgChild70
```

```python
predicted_heights = heights.apply(perdict,'parent_average')
```

```python
notab = heights.with_column("prediction",predicted_heights)
```

```python
notab.show(10)
```

```python
heights.select("parent_average",'child','prediction').scatter('parent_average')
```

```python
def difference(x,y):
    return x-y
pred_errs = heights.apply(difference, 'prediction','child')
heights = heights.with_column("errors",pred_errs)
heights
```

```python
heights.hist("errors", group='Sex')
```

```python
def prediction_smarter(h,s):
    nearby = heights.where('parent_average',are.between(h-0.5,h+0.5))
    nearby_same_sex=nearby.where("Sex",s)
    return np.average(nearby_same_sex.column("child"))
```

```python
prediction_smarter(68,'female')
```

```python
prediction_smarter(68,'male')
```

```python
prediction_smarter_smarter = heights.apply(prediction_smarter,'parent_average','Sex')
```

```python
heights = heights.with_columns("Smarter Prediction", prediction_smarter_smarter)
heights
```
