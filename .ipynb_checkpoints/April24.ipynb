```python
%matplotlib inline
import numpy as np
from datascience import *
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
scores = Table.read_table(".csv/scores_by_section.csv")
scores.show()
```

```python
s = scores.group("Section", np.average)
```

```python
observed_average = s.where("Section", 3)
observed_average = observed_average.get("Midterm average")
```

```python
random_sample = scores.sample(27, with_replacement = False)
np.average(random_sample.column("Midterm"))
```

```python
def random_sample_midterm_avg():
    random_sample = scores.sample(27, with_replacement = False)
    return np.average(random_sample.column("Midterm"))
    
```

```python
sample_averages = make_array()
for i in np.arange(50000):
    sample_averages = np.append(sample_averages, random_sample_midterm_avg())
print(sample_averages)
np.average(sample_averages)
```

```python
len(sample_averages)
```

d = Table().with_column("ds", sample_averages);
d.hist("ds", bins = 50)

```python

sample_averages < observed_average
```

```python
np.count_nonzero(sample_averages <= observed_averages)/50000
```

```python
d.hist(bins=20)
five_percent_point = d.sort(0).column(0).item(2500)
plt.plot([five_percent_point, five_percent_point], [0,0.35], color = 'gold', lw = 2)
plt.title("Area to the left of the gold line: 5%")
plt.scatter(observed_average, -0.01, color = "red", s = 120)
```
