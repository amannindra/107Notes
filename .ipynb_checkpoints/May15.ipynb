```python
for feature in baby.labels:
    values = baby.column(feature)
    mean = np.mean(values)
    sd = np.std(values)
    print()
    print(feature)
    for z in make_array(2,3,4,5):
        chose = baby.where(feature, are.between(mean-z*sd, mean + z*sd))
        proportion = chose.num_rows/baby.num_rows
        percent = round(proportion*100, 2)
        print('Average plus or minus',z,"SD",percent,"%")
```

```python
def standard_units(x):
    #x is an array
    return (x-np.mean(x))/np.std(x)
```

```python
ages = baby.column("Maternal Age")
ages
```

```python
ages_standard_units = standard_units(ages)
ages_standard_units
```

```python
np.mean(ages_standard_units),np.std(ages_standard_units)
```

```python
united = Table.read_table(".csv/united.csv")
```

```python
united
```

```python
united_bins=np.arange(-20,300,10)
united.hist("Delay", bins = united_bins)
```

```python
delays = united.column("Delay")
delay_mean = np.mean(delays)
delay_sd = np.std(delays)
delay_mean, delay_sd
```

```python
def one_sample_mean(sample_size):
    '''Takes a sample form the populatiion of flights and computes its mean of the delay time'''
    sampled_flights = united.sample(sample_size)
    return np.mean(sampled_flights.column("Delay"))
    
```

```python
one_sample_mean(100)
```

```python
def ten_thousand(sample_size):
    means = make_array()
    for i in np.arange(10000):
        mean = one_sample_mean(sample_size)
        means = np.append(means,mean)
    return means
```

```python
q = ten_thousand(100)
```

```python
np.mean(q)
```

```python
Table().with_column("Mena of 100 Flight Delays", q).hist(bins = 20)
print("Population Average", delay_mean)
print("Sampled Mean", np.mean(q))
```

```python
w = ten_thousand(400)
np.mean(w)
Table().with_column("Mena of 100 Flight Delays", w).hist(bins = 20)
print("Population Average", delay_mean)
print("Sampled Mean", np.mean(w))
```

```python
f = ten_thousand(1000)
np.mean(f)
Table().with_column("Mena of 100 Flight Delays", f).hist(bins = 20)
print("Population Average", delay_mean)
print("Sampled Mean", np.mean(f))
```
