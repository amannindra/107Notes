```python
def tvd(dist1, dist2):
    return sum(abs(dist1-dist2))/2
```

```python
obs = tvd(jury.column("Pan"), jury.column("Elg")) 
obs
```

```python
model = make_array(0.15, 0.18, 0.54, 0.12, 0.01)
tvd(sample_proportions(1453, model),jury.column("Elg"))
```

```python
def simulated_tvd():
    return tvd(sample_proportions(1453, model),jury.column("Elg"))
    
```

```python
simulated_tvd()
```

```python
tvds = make_array()
for i in np.arange(10000):
    tvds = np.append(tvds, simulated_tvd())
tvds
```

```python
title = "Simulated TVDS"
bins = np.arange(0,0.05,0.005)
Table().with_column(title, tvds).hist(bins = bins)
print("Observed TVD: " + str(obs))
```
