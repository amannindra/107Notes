```python
game_outcomes = make_array()

for i in np.arange(5):
    game_outcomes = np.append(game_outcomes, one_round(np.random.choice(dice_faces),np.random.choice(dice_faces)))
```

```python
def simulate():
    player1 = np.random.choice(dice_faces)
    player2 = np.random.choice(dice_faces)
    print(f"My roll {player1}, Your roll {player2}")
    return one_round(player1,player2)
```

```python
new_game = make_array()
for i in np.arange(10000):
    new_game = np.append(new_game, simulate())
    
```
