```python
trip = Table.read_table(".csv/trip.csv")
```

```python
trip
```

```python
trip.where("Start Station",are.equal_to,(trip.column("End Station")))
trip
```

```python
def trip_wind(start,end):
    if (start==end):
        return "round print"
    else:
        return "one way"
kins = trip.with_column("Trip Kind", trip.apply(trip_wind, "Start Station", "End Station"))
```

```python
kins.show(5)
```

```python
kins.where("Duration", are.below(600)).pivot("Trip Kind", "Start Station")
```

```python
def one_round(my,your):
    if(my>your):
        return 1
    elif(my < your):
        return -1
    else:
        return 0
    
```

```python
one_round(4,3)
```
