```python
age = 2022-top_movies.column("Year")
```

```python
movies =top_movies.with_column("Age", age)
```

```python
my_bins = make_array(0,5,10,15,25,40,65,100)
my_bins
```

```python
binned_data = movies.bin("Age",bins=my_bins)
binned_data
```

```python
# Another way to create equally spaced bins
```

```python
binned_data
```

```python
binned_data = binned_data.with_column("Percent",((binned_data.column("Age count")/sum(binned_data.column("Age count"))*100).round(1)))
```

```python
binned_data.plot()
```

```python
binned_data.bar()
```
