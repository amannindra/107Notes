```python
observed_values = 709/929
observed_values
```

```python
predicted_proportions = make_array(.75,0.25)
```

```python
sample_proportions(929, predicted_proportions)
```

```python
def purple_floor():
    return sample_proportions(929, predicted_proportions)
purples = make_array()
for i in np.arange(10000):
    new_purple = purple_floor()[0]
    purples = np.append(purples,new_purple)
```

```python
Table().with_column("Percent of Purple Flowers in sample of 929", purples).hist()
```

```python
Table().with_column("Discrepancy in sample of 929 if model is True", purples).hist()
```
